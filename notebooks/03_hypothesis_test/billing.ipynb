# Hypothesis Testing

## Hypothesis 1: Connection Group vs Churn
- H0: Connection Group has no association with churn.
- H1: Connection Group significantly affects churn.

```python
import pandas as pd
from scipy.stats import chi2_contingency

sig = []
not_sig = []

data = df_billing[['Connection_Group', 'target']].dropna()

contingency_table = pd.crosstab(data['Connection_Group'], data['target'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print(" Chi-Square Test: Connection_Group vs Churn")
print("-" * 50)
print("Chi2 Statistic :", chi2)
print("Degrees of Freedom :", dof)
print("P-value :", p)

print("\nContingency Table:\n", contingency_table)
print("\nExpected Frequencies:\n", expected)

alpha = 0.05
feature = "Connection_Group"

if p < alpha:
    print("\n Reject H0: Connection Group significantly affects churn")
    sig.append(feature)
else:
    print("\n Fail to Reject H0: No significant association between Connection Group and churn")
    not_sig.append(feature)
```

## Hypothesis 8: Tenure Years vs Churn
- H0: Tenure distribution is the same for churn and non-churn users.
- H1: Tenure differs between churn and non-churn users.

```python
import pandas as pd
from scipy.stats import ttest_ind

try:
    sig
except NameError:
    sig = []
    not_sig = []

data = df_billing[['Tenure_Years', 'target']].dropna()

churn = data[data['target'] == 1]['Tenure_Years']
non_churn = data[data['target'] == 0]['Tenure_Years']

if len(churn) == 0 or len(non_churn) == 0:
    print(" One group is empty. Check target encoding.")
else:
    t_stat, p = ttest_ind(churn, non_churn, equal_var=False)

    print(" T-Test: Tenure vs Churn")
    print("-" * 50)
    print("T-Statistic:", t_stat)
    print("P-value:", p)

    print("\nMean (Churn):", churn.mean())
    print("Mean (Non-Churn):", non_churn.mean())

    feature = "Tenure_Years"
    alpha = 0.05

    if p < alpha:
        print("\n Reject H0: Tenure mean differs between churn groups")
        sig.append(feature)
    else:
        print("\n Fail to Reject H0: No significant difference in tenure mean")
        not_sig.append(feature)

    print("\n==============================")
    print("SIGNIFICANT FEATURES:", sig)
    print("NON-SIGNIFICANT FEATURES:", not_sig)
    print("==============================")
```

## Hypothesis 9: Price Increase and Total Net Paid vs Churn
- H0: Price increase and total net paid do not differ between churn groups.
- H1: At least one differs between churn groups.

```python
import pandas as pd
from scipy.stats import mannwhitneyu

try:
    sig
except NameError:
    sig = []
    not_sig = []

features = ['price_increase_pct', 'Total_Net_Paid']

for feature in features:
    print(f"\n Mann–Whitney U Test: {feature} vs Churn")
    print("-" * 60)

    data = df_billing[[feature, 'target']].dropna()

    churn = data[data['target'] == 1][feature]
    non_churn = data[data['target'] == 0][feature]

    if len(churn) == 0 or len(non_churn) == 0:
        print(" Skipped: One group is empty")
        continue

    u_stat, p = mannwhitneyu(churn, non_churn, alternative='two-sided')

    print("U-Statistic:", u_stat)
    print("P-value:", p)

    print("Median (Churn):", churn.median())
    print("Median (Non-Churn):", non_churn.median())

    alpha = 0.05

    if p < alpha:
        print(" Reject H0")
        sig.append(feature)
    else:
        print(" Fail to Reject H0")
        not_sig.append(feature)

print("\n==============================")
print("SIGNIFICANT FEATURES:", sig)
print("NON-SIGNIFICANT FEATURES:", not_sig)
print("==============================")
```
